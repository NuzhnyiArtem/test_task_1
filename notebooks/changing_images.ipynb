```python
import albumentations as A
import cv2 as cv
from os import listdir
```

```python
def blur3(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет размытое изображение с уровнем размытия 3.
    Так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "Blur3"
    transform = A.Compose([
        A.Blur(blur_limit=(3, 3), p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output_dir)
    pixel_dropout_2(transformed_image, name, output_dir)


def blur5(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет размытое изображение с уровнем размытия 5.
    Так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "Blur5"
    transform = A.Compose([
        A.Blur(blur_limit=(5, 5), p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output_dir)
    pixel_dropout_2(transformed_image, name, output_dir)


def blur7(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет размытое изображение с уровнем размытия 7.
    Так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "Blur7"
    transform = A.Compose([
        A.Blur(blur_limit=(7, 7), p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output)
    pixel_dropout_2(transformed_image, name, output)


def blur9(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет размытое изображение с уровнем размытия 9.
    Так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "Blur9"
    transform = A.Compose([
        A.Blur(blur_limit=(9, 9), p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output)
    pixel_dropout_2(transformed_image, name, output)


def gausnoise(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет изобржение на которое добавлен Гауссовский шум.
    Так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "GaussNoise"
    transform = A.Compose([
        A.GaussNoise(var_limit=(10.0, 50.0), mean=0, per_channel=True, always_apply=False, p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output)
    pixel_dropout_2(transformed_image, name, output)


def defocus(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет дефокусированное изображение.
    Так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "Defocus"
    transform = A.Compose([
        A.Defocus(radius=(3, 10), alias_blur=(0.1, 0.5), always_apply=False, p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output)
    pixel_dropout_2(transformed_image, name, output)


def mb7(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет изобржение с добавлением размытия в движении с уровнем размытия 7.
    так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "MB7"
    transform = A.Compose([
        A.MotionBlur(blur_limit=7, allow_shifted=True, always_apply=False, p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output_dir + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output)
    pixel_dropout_2(transformed_image, name, output)


def mb9(img, output):
    """Принимает на вход изображение, и путь к папке сохранения фалов,
    сохраняет изобржение с добавлением размытия в движении с уровнем размытия 9.
    так же передает это изображение в функцию pixel_dropout_1 и pixel_dropout_2.
    """
    name = "MB9"
    transform = A.Compose([
        A.MotionBlur (blur_limit=9, allow_shifted=True, always_apply=False, p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + file, transformed_image)
    pixel_dropout_1(transformed_image, name, output)
    pixel_dropout_2(transformed_image, name, output)


def pixel_dropout_1(img, name, output):
    """Принимает на вход изображение имя функции вызвавшей данную функцию и путь к папке сохранения файлов.
    Сохраняет изображение с установленным значением 0 для некоторых пикселей с частотой равной 0.01
    """
    namepd1 = "PD1"
    transform = A.Compose([
        A.PixelDropout(dropout_prob=0.01, per_channel=False, drop_value=0, mask_drop_value=None, always_apply=False, p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + namepd1 + "_" + file, transformed_image)


def pixel_dropout_2(img, name, output):
    """Принимает на вход изображение имя функции вызвавшей данную функцию и путь к папке сохранения файлов.
    Сохраняет изображение с установленным значением 0 для некоторых пикселей с частотой равной 0.03
    """
    namepd2 = "PD2"
    transform = A.Compose([
        A.PixelDropout(dropout_prob=0.03, per_channel=False, drop_value=0, mask_drop_value=None, always_apply=False,
                       p=1)
    ])
    transformed = transform(image=img)
    transformed_image = transformed["image"]
    cv.imwrite(output + "\\" + name + "_" + namepd2 + "_" + file, transformed_image)


path_image = r""  # Путь к папке с изображениями
output_dir = r""  # Путь к папке куда будут сохраняться обработанные изображения

files = listdir(path_image)

for file in files:
    image = cv.imread(path_image + "\\" + file)
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    blur3(image, output_dir)
    blur5(image, output_dir)
    blur7(image, output_dir)
    blur9(image, output_dir)
    defocus(image, output_dir)
    gausnoise(image, output_dir)
    mb7(image, output_dir)
    mb9(image, output_dir)
```
